Как написано в лабе тупо сдвигаем график по желтой линии неона

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

fig, ax = plt.subplots()

fi=[1700,1850,2010,2150,2300,2410,2520,2608,2690,2820,2990,3120,3150,3230, 3310, 3340,3420, 3480, 3500]
Lambda= [4700,4900,5100,5300,5500,5700,5900,6100,6300,6700,7300,7900,8100,8500,8900,9100,9500,9900,10000]
fi_error= 10
Lambda_error = 1

fi_new = [0]*len(fi)

for i in range (len(fi)):
    fi_new[i] = fi[i] -15.7

z = np.polyfit(fi_new, Lambda, 4)
f = np.poly1d(z)
f_new = np.linspace(fi_new[0], fi_new[-1], 100)
l_new = f(f_new)

print(f.c) 


plt.xlabel("Theta, grad")
plt.ylabel("lambda, angstrem")
plt.scatter(fi_new, Lambda, s = 1)
plt.plot(f_new, l_new, linewidth = 1)
ax.minorticks_on()
ax.grid(which='major', linewidth = 0.75)
ax.grid(which='minor', linewidth = 0.5)
plt.grid(True) 
plt.errorbar(fi_new, Lambda,yerr = Lambda_error, xerr = fi_error,fmt='none',ecolor = 'black',elinewidth = 1)

fig.savefig('grad.png', dpi = 1000)
```
